# Exercise 7.2: Wordcount

```python
import re
file='shakespeare.txt'
'''
(a) Find the 20 most common words
(b) How many unique words are used?
(c) How many words are used at least 5 times?
(d) Write the 200 most common words, and their counts, to a file.
'''
#(a) Find the 20 most common words
words={}
with open(file) as ifile:
    #for each line in file
    for line in ifile.readlines():
        #split line to get words
        line=line.strip().split()
        for word in line:
            #check if the word does not exist in the dictionary
            #get only the word without pounctiations 
            word=re.sub("[^\w\s]", "", word)
            #if word=="thou":
            #    print(line)
            if word not in words:
                #initilize its frequency to 1
                words[word]=1
            else:
                #increment its frequency by 1
                words[word]+=1
    #sort words by their frequnecies
    sorted_values=[k for k in {k: v for k, v in sorted(words.items(), key=lambda item: item[1])}]
    #top 50 words
    top_50_words=[v for v in sorted_values[len(sorted_values)-50:]]
    print("a) TOP 50 WORDS:\t",top_50_words)

#(b) How many unique words are used?
print("\n\nb) There are", len(words), " unique words.")

#(c) How many words are used at least 5 times?
print("\n\nc) There are", len([c for c in words if words[c]>3]), " words that are used at least 5 times.")

#(d) Write the 200 most common words, and their counts, to a file.
with open ("200_words.txt","w") as ofile:
    top_200_words=[v for v in sorted_values[len(sorted_values)-200:]]
    for word in top_200_words:
        ofile.write(word+"\n")
```

```python
'''
(a) Data generation
Write a function that takes three integers, n, a and b and a filename and writes to the file a list
with n random integers between a and b.
'''
import random
def DataGeneration(n,a,b,filename):
    with open(filename,"w") as ofile:
        for i in range(n):
            ofile.write(str(random.randint(a,b))+"\n")
DataGeneration(20,1,24,"Exercise 7.4: Sum of lists.txt")
```
