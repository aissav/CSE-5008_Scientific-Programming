# Exercise 11.1: Plotting a function

Plot the function

$f(x) = sin^2(x − 2)e^{-x^2}$

over the interval [0, 2]. Add proper axis labels, a title, etc.

```python
def EX10_2(x):  
    #$f(x) = sin^2(x − 2)$
    sin_square=np.power(np.sin(x-2), 2)
    #e^{-x^2}
    exp=np.exp(-1*(np.power(x,2)))
    return 1 * sin_square * exp

#generate k x values between 0 and 2
k=25
X_values = np.linspace(0, 2, k)

#calculate Y on the generated X values between 0 and 2
Y_Values = EX10_2(X_values)

#plot the function
plt.plot(X_values, Y_Values)
plt.xlabel('X')
plt.ylabel('F(x)')
plt.title('Ex 11.1')

plt.show()
```

# Exercise 11.3: Histogram and density estimation

Generate a vector z of 10000 observations from your favorite exotic distribution. 

Then make a plot that shows a histogram of z (with 25 bins), along with an estimate for the density, using a 

Gaussian kernel density estimator (see scipy.stats). See Figure 2 for an example plot.

```python
10000/25
```

```python
bins=25
n_values=10000
#z values
z = np.random.randn(n_values)


#using a Gaussian kernel density estimator
gaussiane_kernel = stats.gaussian_kde(z)

estimated_values = gaussiane_kernel.pdf(np.linspace(-20, 20, 100))
plt.hist(z, bins=bins, density=True)
plt.plot(np.linspace(-20, 20, 100), estimated_values, '#000000', lw=1)
plt.show()
```
